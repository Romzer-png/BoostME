```python
import requests
import os
from dotenv import load_dotenv
```

```python
load_dotenv()
API_KEY = os.getenv("GOOGLE_API_KEY")

```

```python
url = "https://www.googleapis.com/youtube/v3/search"

params = {
    "part": "snippet",
    "q": "fitness",
    "type": "video",
    "maxResults": 5,
    "regionCode": "FR",
    "key": API_KEY
}
```

```python
response = requests.get(url, params=params)
response
```

```python
data = response.json()

video_ids = []
for item in data["items"]:
    video_ids.append(item["id"]["videoId"])

video_ids
```

```python
url_videos = "https://www.googleapis.com/youtube/v3/videos"

params_videos = {
    "part": "snippet,statistics,contentDetails",
    "id": ",".join(video_ids),
    "key": API_KEY
}

response_videos = requests.get(url_videos, params=params_videos)
print(response_videos.status_code)

videos_data = response_videos.json()
```

```python
rows = []

for video in videos_data["items"]:
    rows.append({
        "video_id": video["id"],
        "title": video["snippet"]["title"],
        "channel": video["snippet"]["channelTitle"],
        "published_at": video["snippet"]["publishedAt"],
        "views": int(video["statistics"].get("viewCount", 0)),
        "likes": int(video["statistics"].get("likeCount", 0)),
        "comments": int(video["statistics"].get("commentCount", 0)),
    })

rows
```

```python
import pandas as pd

df = pd.DataFrame(rows)
df
```

```python
df.to_csv("market_fitness_fr.csv", index=False)
print("CSV sauvegard√© ‚úÖ")
```
