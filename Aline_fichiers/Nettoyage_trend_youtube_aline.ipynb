```python
import gdown

file_id = "1FPUEC1YzdyRyrLV9Ufmr39CgI5aclN_T"  # Remplace par ton ID
url = f"https://drive.google.com/uc?id={file_id}"
output = "trend_videos_youtube_aline.csv"  # Nom local pour le CSV
gdown.download(url, output, quiet=False)
```

```python
import pandas as pd

df = pd.read_csv("trend_videos_youtube_aline.csv")
df.head()
```

```python
# V√©rifier les valeurs manquantes
print(df.isnull().sum())

# V√©rifier les doublons sur l'ID vid√©o
print("Doublons Video ID :", df.duplicated(subset="Video ID").sum())

# Supprimer les doublons si n√©cessaire
df = df.drop_duplicates(subset="Video ID")
```

```python
# V√©rifier les doublons sur l'ID vid√©o
print("Doublons Video ID :", df.duplicated(subset="Video ID").sum())
```

```python
# normaliser les formats
# transofrmer les dur√©es (ici min/sec) en secondes (plus simple pour l'analyse
import re

def duration_to_seconds(duration_str):
    # Convertit "minutes:seconds" en secondes
    match = re.match(r"(\d+):(\d+)", duration_str)
    if match:
        minutes, seconds = match.groups()
        return int(minutes) * 60 + int(seconds)
    return 0

df["Dur√©e (s)"] = df["Dur√©e (min:sec)"].apply(duration_to_seconds)
```

```python
# ajout de colonnes de dates et heures de publication pour analyser les meilleurs moments pour publier
df["Date de publication"] = pd.to_datetime(df["Date de publication"], format="%d/%m/%Y %H:%M")
df["Ann√©e"] = df["Date de publication"].dt.year
df["Mois"] = df["Date de publication"].dt.month
df["Jour"] = df["Date de publication"].dt.day
df["Heure"] = df["Date de publication"].dt.hour
df["Jour de la semaine"] = df["Date de publication"].dt.day_name()
```

```python
# calcuker le taux d'engagement pour voir en un coup d'oeil les vid√©os ayant re√ßu le plus d'int√©ractions
df["Engagement total"] = df["Likes"] + df["Commentaires"]
df["Taux d'engagement (%)"] = (df["Engagement total"] / df["Vues"].replace(0, 1)) * 100
```

```python
# remplacer les valeurs manquantes dans genre par "Inconnu"

print(df["Genre"].value_counts())

# Remplacer les valeurs manquantes ou "Inconnu"
df["Genre"] = df["Genre"].fillna("Inconnu")
```

```python
# ajout d'une colonne[Descritpion_clean] pour pousser les analyse par mots cl√©s (NLP, hashtags, mots cl√©s)
# Nettoyer les descriptions
df["Description_clean"] = df["Description"].str.replace(r'\n', ' ', regex=True)
df["Description_clean"] = df["Description_clean"].str.strip()
```

```python
df.to_csv("trend_youtube_aline_clean.csv", index=False)
```
